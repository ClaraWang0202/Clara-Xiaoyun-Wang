```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
path='/Users/clarawang/Dropbox/Python case/'
data=pd.read_excel(path+'不满意率分析0415.xlsx')
print("数据规模:")
print(data.shape)
```

```python
data.head(6)
```

```python
data['评价标签']=data['顾客配送评价标签'].str.split('|')
#create new column named (评价标签)
```

```python
data.head()
```

```python
# 所有的评价标签
total_list=[]
for i in data['评价标签']:
    for j in i:
        total_list.append(j)
set(total_list)
```

```python
#判断是否仪表不整
def judge_1(x):
    total = 0
    for i in x['评价标签']:
        if i == '仪表不整':
            total += 1
    return total
data['仪表不整'] = data.apply(judge_1, axis=1)

# 判断其他
def judge_2(x):
    total = 0
    for i in x['评价标签']:
        if i == '其他':
            total += 1
    return total
data['其他'] = data.apply(judge_2, axis=1)

# 判断少餐/洒餐
def judge_3(x):
    total = 0
    for i in x['评价标签']:
        if i == '少餐/洒餐':
            total += 1
    return total
data['少餐/洒餐'] = data.apply(judge_3, axis=1)

# 判断是否态度不好
def judge_4(x):
    total = 0
    for i in x['评价标签']:
        if i == '态度不好':
            total += 1
    return total
data['态度不好'] = data.apply(judge_4, axis=1)

# 判断提前点送达
def judge_5(x):
    total = 0
    for i in x['评价标签']:
        if i == '提前点送达':
            total += 1
    return total
data['提前点送达'] = data.apply(judge_5, axis=1)

# 判断送达不通知
def judge_6(x):
    total = 0
    for i in x['评价标签']:
        if i == '送达不通知':
            total += 1
    return total
data['送达不通知'] = data.apply(judge_6, axis=1)

# 判断送达超时
def judge_7(x):
    total = 0
    for i in x['评价标签']:
        if i == '送达超时':
            total += 1
    return total
data['送达超时'] = data.apply(judge_7, axis=1)

# 判断送错餐品
def judge_8(x):
    total = 0
    for i in x['评价标签']:
        if i == '送错餐品':
            total += 1
    return total
data['送错餐品'] = data.apply(judge_8, axis=1)

# 判断骚扰威胁
def judge_9(x):
    total = 0
    for i in x['评价标签']:
        if i == '骚扰威胁':
            total += 1
    return total
data['骚扰威胁'] = data.apply(judge_9, axis=1)
```

```python
# 异常值处理 对数值类型字段: [骑手接单时长, 到店时长, 到店等餐时长, 送达时长]进行标准化，
std_data =  data[['骑手接单时长', '到店时长', '到店等餐时长', '送达时长']]
```

```python
std_data.head()
```

```python


# 骑手接单时长_std
std_data['骑手接单时长_std'] = (std_data['骑手接单时长'] - std_data['骑手接单时长'].mean()) / std_data['骑手接单时长'].std()
temp_index = std_data[(std_data['骑手接单时长_std']>3) | (std_data['骑手接单时长_std']<-3)].index
data.drop(index=temp_index, axis=0, inplace=True)

# 到店时长_std
std_data['到店时长_std'] = (std_data['到店时长'] - std_data['到店时长'].mean()) / std_data['到店时长'].std()
temp_index = std_data[(std_data['到店时长_std']>3) | (std_data['到店时长_std']<-3)].index
data.drop(index=temp_index, axis=0, inplace=True)

# 到店时长_std
std_data['到店等餐时长_std'] = (std_data['到店等餐时长'] - std_data['到店等餐时长'].mean()) / std_data['到店等餐时长'].std()
temp_index = std_data[(std_data['到店等餐时长_std']>3) | (std_data['到店等餐时长_std']<-3)].index
data.drop(index=temp_index, axis=0, inplace=True)

# 送达时长_std
std_data['送达时长_std'] = (std_data['送达时长'] - std_data['送达时长'].mean()) / std_data['送达时长'].std()
temp_index = std_data[(std_data['送达时长_std']>3) | (std_data['送达时长_std']<-3)].index
data.drop(index=temp_index, axis=0, inplace=True)
```
